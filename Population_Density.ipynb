## Population Density

```python
population = xr.open_dataset('/home/cccr/supriyo/abir/population/gpw_v4_population_density_rev11_2pt5_min.nc')
```

```python
population.keys()
po = population["Population Density, v4.11 (2000, 2005, 2010, 2015, 2020): 2.5 arc-minutes"]
po.values.shape
```

```python
lon_m = po.longitude
lat_m = po.latitude
```

## Population Count

```python
pop_count = xr.open_dataset('gpw_v4_population_count_rev11_15_min.nc')
```

```python
pop_count['pop_count'] = pop_count['Population Count, v4.11 (2000, 2005, 2010, 2015, 2020): 15 arc-minutes']
```

#### India and China Population

```python
india = pop_count.sel(longitude=slice(72,92),latitude=slice(33,8)).sum(dim='latitude').sum(dim='longitude')
china = pop_count.sel(longitude=slice(103,122),latitude=slice(40,22)).sum(dim='latitude').sum(dim='longitude')
```

##### World Population

```python
world = pop_count.pop_count[0:5,:,:].sum(dim='latitude').sum(dim='longitude')
```

Percentage of total population

```python
india_per = 100*india.pop_count[0:5]/world
china_per = 100*china.pop_count[0:5]/world
```

```python
india_china     = india.pop_count[0:5] + china.pop_count[0:5]
india_china_per = 100*india_china/world
```

## Plotting

```python
cmap1 = nclcmaps.cmap('WhiteBlueGreenYellowRed')
cmap1 = get_cmap('YlOrBr')
cmap1 = get_cmap('Blues')

# cmap1 = nclcmaps.cmap('precip2_17lev')
img_extent =[40, 150, -20, 60]

wu_lon, wu_lat       = 114.30, 30.59
delhi_lon, delhi_lat = 77.23, 28.61
nay_lon, nay_lat     = 96.07, 19.76


fig=plt.figure(figsize=[24,8])
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent(img_extent, ccrs.PlateCarree())

fname = '/home/cccr/supriyo/abir/OMI/NO2/analysis/india_shape_file/new_shape/India_Boundary.shp'
shape_feature = ShapelyFeature(shpreader.Reader(fname).geometries(),
                                ccrs.PlateCarree(), facecolor='none',edgecolor='k',linewidth=0.8,linestyle="-",alpha=0.8)
ax.add_feature(shape_feature)
c = ax.contourf(lon_m, lat_m, po[4,:,:], extent=img_extent, cmap=cmap1,
                levels=np.arange(200,2000,500),extend='both',transform=ccrs.PlateCarree())

plt.colorbar(c,ax=ax,pad=0.15,shrink=0.55,orientation='vertical')
ax.coastlines()
ax.set_title('Population Density as of 2020 (persons/sq. km)')
ax.set_xticks([60,90,120,150], crs=ccrs.PlateCarree())
ax.set_yticks([ -20,-10, 0,10, 20,30,40, 50, 60], crs=ccrs.PlateCarree())
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)
ax.tick_params(axis='both',labelsize=14)

BORDER_ZORDER = 1110
states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='10m',
        facecolor='none')

ax.add_feature(states_provinces, edgecolor='black', linestyle=":",zorder=BORDER_ZORDER,alpha=0.3)
    
ax.add_feature(cfeature.BORDERS, linestyle='-')



plt.plot(wu_lon, wu_lat, color='r', marker='o', markersize=6,
         transform=ccrs.Geodetic(),
         )

plt.plot(delhi_lon, delhi_lat,
         color='r', marker = 'o',markersize=6,
         transform=ccrs.PlateCarree(),
         )

#plt.plot(nay_lon, nay_lat, color='yellow', marker='o', markersize=6,
#         transform=ccrs.Geodetic(),
#         )

plt.text(wu_lon - 3, wu_lat +6, 'Wuhan',
         horizontalalignment='right',color='r',
         transform=ccrs.Geodetic())

plt.text(delhi_lon - 3, delhi_lat + 6, 'Delhi',color='r',
         horizontalalignment='left',
         transform=ccrs.Geodetic())

rect1 = patches.Rectangle((70,8),22,25,linewidth=1,edgecolor='k',facecolor='None')
#rect2 = patches.Rectangle((92,14),14,10,linewidth=1,edgecolor='k',facecolor='None')
rect3 = patches.Rectangle((103,22),19,18,linewidth=1,edgecolor='k',facecolor='None')

ax.add_patch(rect1)
#ax.add_patch(rect2)
ax.add_patch(rect3)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax.text(0.15, 0.15, 'a', transform=ax.transAxes, fontsize=20,
        verticalalignment='top', bbox=props)


plt.savefig('/home/cccr/supriyo/abir/OMI/plots_COVID-19_pollution/population.png',dpi=300)
```
